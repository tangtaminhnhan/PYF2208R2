## bai3
nhap vao tien luong theo gio va so gio trong 1 thang va tien thuong
tong thu nhap=tien luong theo gio *so gio + tien thuong

```python
tien_luong_theo_gio=int(input("nhap vao tien luong:"))
so_gio=int(input("nhap vao so gio:"))
tien_thuong=int(input("nhap vao tien thuong:"))
print(f" tong thu nhap la {tien_luong_theo_gio*so_gio+tien_thuong}")
```
