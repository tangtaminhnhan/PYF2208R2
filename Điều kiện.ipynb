## Câu lệnh if

```python
mark=float(input("nhap vao diem:"))
if mark >=8.5:
    print("ban duoc tuyen thang")
```

```python
mark=float(input("nhap diem:"))
if mark >=8.5:
    print("ban duoc tuyen thang")
else:
    print("ban chưa du diem")
```

### Thực hành:
Viết chương trình nhập vào môn học hiện tại. Nếu môn học là python thì in ra "I'm studying python"

```python
nhap=input("nhap mon hoc hien tai")
if nhap=="I'm studying python":
    print("I'm studying python")
```

## Câu lệnh if..else

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
n=int(input("nhap voa so nguyen"))
if n%2==9:
    print("so chan")
else :
    print("so le")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
chuoi=input("nhap vao chuoi")
if int(chuoi.count("python"))>0:
    print("co")
else:
    print("khong")
```
